## Let us attach to it all the parts according to D3.1  and D3.2

**We see that a Title must have some string value according to W3C standards. 
This demonstrates that each CUDS carries on it the entire ontological axioms (relation) that it is expected to have.**

Let us define the various parts of the DomeDataSet entity: 

```python
title=domeo.Title(string_value="This is the title of the DOME Data Set")
```

```python
keywords=domeo.SyntacticKeyword(string_value="DOME4.0, stress, Cu")
```

```python
semantic_keywords=domeo.SemanticKeyword()
```

```python
stress_keyword=emmo.Stress() 
```

*We semantically attach a concept, Stress, to the semantic keywords! i.e., stress here is an ontology concept which may contain further knowledge.*

```python
semantic_keywords.connect(stress_keyword, rel=domeo.hasPart)
```

### The `connect` creates an arch - or relation between the semantic keywors and the concept (a triplet):

```python
semantic2dot(semantic_keywords, stress_keyword)
```

**Instead of the less semantic (in fact syntactic) Ontology Annotations, which are meant for human consmption and not understandable by a machine (algorithm), hence not fully interoperable, we use in CUDS a proper seamntic entity for textual elements like Description, and title etc. We also use an OWL Data Proprty (`value`, `string_value`, etc) to provide Literals:**

```python
description=domeo.Description(string_value="This is the description of this data set")
```

```python
issued_date=domeo.IssuedDate(date_value="2023-08-16T10:20:15-05:00")
```

```python
publisher=domeo.Publisher(iri="https://open-research-europe.ec.europa.eu")
```

```python
creator=domeo.Creator(iri="https://como.ceb.cam.ac.uk/people/ab349/")
```

```python
GPL=domeo.License(iri="https://www.gnu.org/licenses/gpl-3.0.html")
```

```python
license_txt=domeo.Description(string_value="The GNU General Public License is a free, copyleft license for software and other kinds of works.")
```

```python
GPL.connect(description, rel=domeo.hasPart)  # we can easilly add more semantic knwoledge as triples. 
```

```python
tier=domeo.Tier1() # signify Tier 1 semantic intop, i.e., we only have a link to the data and keywords etc., as shown above. See D3.6.
```

## Now Compose the Data Set by creating the connections (archs) to all nodes


```python
data_set.connect(keywords, rel=domeo.hasPart)
```

```python
data_set.connect(semantic_keywords, rel=domeo.hasPart)
```

```python
data_set.connect(description, rel=domeo.hasPart)
```

```python
data_set.connect(issued_date, rel=domeo.hasPart)
```

```python
data_set.connect(title, rel=domeo.hasPart)
```

```python
data_set.connect(publisher, rel=domeo.hasPart)
```

```python
data_set.connect(tier, rel=domeo.hasPart)
```

```python
data_set.connect(creator, rel=domeo.hasPart)
```

```python
data_set.connect(GPL, rel=domeo.hasPart)
```

```python
# of course we could have done much more smartly in one go :-)

for i in [title, keywords, semantic_keywords, description, issued_date, publisher, creator, GPL, tier]:
    data_set.connect(i, rel=domeo.hasPart)
```

As a quick demo, we show how we can query and find this data set using SPARQL

```python
res = sparql(
        f"""SELECT ?subject ?predicate ?object WHERE {{
    ?subject ?predicate ?object .
    FILTER (
        CONTAINS(STR(?subject), "Dome") ||
        CONTAINS(STR(?predicate), "Dome") ||
        CONTAINS(STR(?object), "Dome")
    )
    }}
    """)
```

```python
len(res)
```

```python
print(res.serialize(format="json"))
```

```python
for i in res:
    print(i)
```

```python
# here is a little utility to unpack the result
def res2list(R, as_json=False):
    results_list = []

    for row in R:
        # Using a dictionary comprehension to get the desired structure
        row_dict = {key: str(value) for key, value in zip(['s', 'p', 'o'], row)}
        results_list.append(row_dict)

    if as_json:
        return json.dumps(results_list, indent=4)
    
    return results_list
```

```python
res=res2list(res)
```
