find all triplets of any class which have crustall in teh iri. 

```python
lt = sparql(
    f"""SELECT ?s ?p ?o WHERE {{
        ?s rdf:type/rdfs:subClassOf* ?type  .
        FILTER(CONTAINS(STR(?s), "CRYS"))
    }}
    """)
```

get all the triples of each entity... an overkill, but to show it can be done!

```python
for i in lt:
    print(core_session.get(i[0]).triples)
```

fiind all entities which have mater in their iri! 

```python
lt = sparql(
    f"""SELECT ?s ?p ?o WHERE {{
        ?s rdf:type/rdfs:subClassOf* ?type  .
      FILTER(CONTAINS(LCASE(STR(?s)), LCASE("mater")))
    }}
    """)
```
